```python
import socket
import time
import numpy as np
import random
import json

def generateJson():
    my_dict = {'BTC':random.uniform(39000, 60000), 
               'ETH':random.uniform(1500, 3000), 
               'GALA':random.uniform(0, 1), 
               'People':random.uniform(0, 1)}

    my_dict = json.dumps(my_dict)
    crypt = listToString(my_dict)
    return crypt

def listToString(inputList): 
    outputStr = '' 
    for n in inputList: 
        outputStr += str(n) + ''

    return outputStr.rstrip() 

sock = socket.socket() 
sock.bind(('',9002)) 
sock.listen(1) 
conn, addr = sock.accept()

n = 60 
i = 0 
while i <= n: 
    try: 
        json_to_send = generateJson() + '\n'
        print(json_to_send)
        conn.send(json_to_send.encode()) 
    except Exception as e: 
        print('Exception: ', e) 
        sock.close() 
        break
    finally: 
        i = i+1 
        time.sleep(5)
        
```
